## Fintech: How Computer Vision can improve payment

In Nigeria's booming fintech space, computer vision can enhance payment systems in several ways:

- **Biometrics**: Facial recognition and fingerprint scanning can enable secure, contactless authentication for mobile banking and ATM transactions.
- **QR-Code Scanning**: Vision-based QR code readers allow fast, low-cost payments in markets, transport, and retail — even on low-end devices.
- **Swipe and Pay**: Gesture recognition can enable futuristic interfaces where users swipe or tap in front of a camera to authorize payments, reducing physical contact and hardware dependency.

These innovations can improve accessibility, reduce fraud, and streamline user experience in Nigeria’s mobile-first economy.

## Healthtech: How Computer Vision can expand access to healthcare

Computer vision can empower healthtech startups in Nigeria to reach underserved populations by:

- **Remote Diagnostics**: Vision models can analyze skin conditions, eye scans, or X-rays via mobile apps, enabling early detection without hospital visits.
- **Medication Recognition**: Apps can identify pills and dosage using camera input, helping patients manage prescriptions safely.
- **Telemedicine Triage**: Facial analysis can detect signs of pain, fatigue, or distress during virtual consultations, improving care quality.
- **Health Kiosks**: Vision-enabled kiosks in rural areas can guide users through self-checkups and connect them to remote doctors.

These tools reduce barriers to care and bring medical expertise closer to communities in need.

```python
# ==========================
# Q3 — Image Restoration Code
# ==========================

import cv2
import matplotlib.pyplot as plt

# Load corrupted image
img_path = "/content/Corrupted Student Picture.png"
img = cv2.imread(img_path)

# Function to adjust contrast & brightness
def adjust_contrast_brightness(image, alpha=1.0, beta=0):
    # alpha = contrast (1.0 = original, >1 higher, <1 lower)
    # beta = brightness (0 = original, + brighter, - darker)
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# Trying some values
alpha = 0.8   # contrast
beta = 40     # brightness
restored = adjust_contrast_brightness(img, alpha, beta)

# Show original vs restored
plt.figure(figsize=(12,6))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("Corrupted Image")
plt.subplot(1,2,2); plt.imshow(cv2.cvtColor(restored, cv2.COLOR_BGR2RGB)); plt.title(f"Restored (alpha={alpha}, beta={beta})")
plt.show()

print(f"Best restoration values: alpha={alpha}, beta={beta}")
```
